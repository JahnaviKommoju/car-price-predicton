```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
A=OneHotEncoder(handle_unknown='ignore',sparse_output=True)
A.fit(x[['Company','Model','Fuel','Gear']])
```

```python
 A.categories_
```

```python
column_trans= make_column_transformer((OneHotEncoder(categories=A.categories_),['Company','Model','Fuel','Gear']),remainder='passthrough')
column_trans
```

```python
LR=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,LR)
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
Sc=[]
for c in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=c)
    LR=LinearRegression()
    pipe=make_pipeline(column_trans,LR)
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    Sc.append(r2_score(y_test,y_pred))
     
```

```python
import numpy as np
d=np.argmax(Sc)
print(d)  #random_state no
print(Sc[d]) #accuracy of the particular random_state
```
